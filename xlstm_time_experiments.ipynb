```python
import pandas as pd
import numpy as np
def plot_preds(preds, num_steps):
    df_pred_vs_target = pd.DataFrame(
        {
            'target': preds[num_steps]['true'].reshape(num_steps, ),
            'pred': preds[num_steps]['predicted'].reshape(num_steps, )
        }
    )
    df_pred_vs_target['step'] = np.arange(1, num_steps + 1)
    df_pred_vs_target.plot(x='step', y=['pred', 'target'], figsize=(18, 9))
```

```python
plot_preds(experiment_pipeline.predictions, 7)
```
